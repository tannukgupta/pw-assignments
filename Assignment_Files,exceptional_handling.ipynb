## 9. What is the difference between multithreading and multiprocessing?
- **Multithreading:**

  - Runs multiple threads within the same process.

  - All threads share the same memory space.

  - Good for I/O-bound tasks (e.g., reading files, network calls).

  - Affected by Python’s Global Interpreter Lock (GIL), so true parallelism is limited.

  - Lower memory usage and faster context switching


Example:

    import threading

    def print_numbers():
        for i in range(5):
            print("Thread:", i)

    t1 = threading.Thread(target=print_numbers)
    t2 = threading.Thread(target=print_numbers)

    t1.start()
    t2.start()
Use Case: Downloading files, handling multiple user requests, reading/writing from files or web

- **Multiprocessing:**

  - Runs multiple processes, each with its own memory.

  - Suitable for CPU-bound tasks (e.g., heavy computations, data processing).

  - Not affected by the GIL — allows true parallelism on multiple CPU cores.

 -  Uses more memory, but achieves better performance for CPU-intensive tasks.

 -  Slower context switching compared to threads.

Example:

    import multiprocessing

    def print_numbers():
        for i in range(5):
            print("Process:", i)

    p1 = multiprocessing.Process(target=print_numbers)
    p2 = multiprocessing.Process(target=print_numbers)

    p1.start()
    p2.start()

Use Case: Image processing, large data computations, parallel simulations.

## 11. What is memory management in Python?
- Memory management in Python is the process by which Python handles the allocation and deallocation of memory to store variables, objects, and data structures during program execution.

- **Mechanisms:**

**Automatic Memory Allocation:**
Python automatically allocates memory when you create variables or objects.

**Reference Counting:**
Each object keeps track of how many references point to it. When the count reaches zero, it's deleted.

**Garbage Collection:**
Python uses a garbage collector to clean up memory used by objects that are no longer needed, especially those involved in circular references.

**Private Heap and PyMalloc:**
Python maintains its own private memory heap and uses an internal memory manager (PyMalloc) for efficiency.

Example:

    import sys

    a = [1, 2, 3]
    b = a  # both a and b point to the same list

    print(sys.getrefcount(a))  # Shows the reference count for object 'a'

    del a  # deletes one reference

    # object still exists because 'b' refers to it

Python manages memory automatically using reference counting and garbage collection, so developers don’t need to manually allocate or free memory

## 15. How does Python's garbage collection system work?

- Python’s garbage collection system automatically frees up memory by removing objects that are no longer needed.

How it works:

**Reference Counting:**

Every object in Python keeps track of how many references point to it.

When the reference count becomes zero, the object is deleted.

**Garbage Collector (GC):**

Handles complex cases like circular references (e.g., objects referring to each other).

Runs automatically in the background using the gc module.

    import gc

    class Node:
        def __init__(self, name):
            self.name = name
            self.ref = None  # Can point to another object

    # ---- Reference Counting ----
    a = Node("A")
    b = a  # Reference count of 'a' is now 2
    del a  # Reference count becomes 1
    del b  # Reference count becomes 0 → object is deleted automatically

    # ---- Circular Reference (Garbage Collector needed) ----
    x = Node("X")
    y = Node("Y")

    x.ref = y
    y.ref = x  # Creates a circular reference

    del x
    del y

    # At this point, x and y still exist in memory due to the cycle
    # Force garbage collection:
    gc.collect()

This example shows how Python first tries reference counting, and if that fails (due to cycles), it uses the garbage collector.


Explanation:
The first part (a, b) shows reference counting – when count drops to 0, the object is deleted.

The second part (x, y) shows a circular reference – even after deleting the variables, memory isn't freed until gc.collect() is called.

## 18. What is the difference between os.fork() and multiprocessing in Python?
- Difference between os.fork() and multiprocessing in Python?

### **os.fork()**

**Abstraction Level:**
os.fork() is a low-level system call used to create a new process by duplicating the current one.

**Platform Support:**
It only works on Unix-like systems (Linux, macOS). It is not available on Windows.

**Memory Handling:**
Memory is handled manually using copy-on-write, which can be complex to manage and prone to errors.

**Process Creation:**
A new process is created using the os.fork() system call, which duplicates the current process.

**Communication:**
Inter-process communication (IPC) must be done manually, often using pipes or shared memory.

**Ease of Use:**
It is considered complex and harder to use, especially for beginners.

**Use Cases:**
Typically used in system-level scripts, background services, and low-level operations.

**Parent/Child Execution:**
After forking, both parent and child processes continue execution from the same line of code, making it harder to manage logic flow.


### **multiprocessing (Python module)**

**Abstraction Level:**
multiprocessing is a high-level module provided by Python for working with processes.

**Platform Support:**
It is cross-platform and works on Windows, Linux, and macOS.

**Memory Handling:**
It handles memory automatically and in isolated environments, reducing errors and making it safer.

**Process Creation:**
New processes are created using the Process class, where you define a target function.

**Communication:**
Communication is easier using built-in tools like Queue, Pipe, and Manager.

**Ease of Use:**
It is simple and beginner-friendly, with a clean API for parallel processing.

**Use Cases:**
Best suited for data processing, concurrent execution, and CPU-bound tasks in Python programs.

**Parent/Child Execution:**
The child process starts by executing a separate function, making the flow of execution easier to control and understand

## 23. What are the challenges associated with memory management in Python?
- Challenges Associated with Memory Management in Python

**Garbage Collection Overhead**

Python uses automatic garbage collection to free unused memory. While convenient, it can cause performance overhead if garbage collection runs too often, especially in large applications.

**Reference Cycles**

When two or more objects reference each other, they form a cycle that the default reference counting mechanism cannot handle. Python’s garbage collector can detect cycles, but this adds complexity and may delay memory release.

**Memory Leaks**

Even with garbage collection, memory leaks can occur when unused objects are kept in memory due to lingering references (e.g., in global variables, caches, or data structures).

**Fragmentation**

Python’s memory allocator may cause fragmentation, where free memory is scattered in small chunks, making it harder to reuse efficiently. This can increase overall memory usage.

**Large Object Retention**

Python may not return freed memory to the operating system immediately, especially for large objects, leading to high memory consumption during long-running programs.

**External Resource Managemen**t

Memory for external resources (like file handles or database connections) is not always handled by Python’s garbage collector. These require explicit closing or cleanup

**Example:**

    import gc

    # 1. Garbage Collection Overhead
    gc.collect()  # Manually trigger garbage collection (may cause pause)

    # 2. Reference Cycle
    class Node:
        def __init__(self):
            self.ref = None

    a = Node()
    b = Node()
    a.ref = b
    b.ref = a  # Creates a reference cycle

    # 3. Memory Leak
    data = []
    for i in range(100000):
        data.append("Leak!")  # Keeps growing without clearing

    # 4. Fragmentation (hard to show directly, but creating & deleting varying objects can cause it)
    temp_objs = [[0] * i for i in range(1, 1000)]
    del temp_objs  # Memory may remain fragmented

    # 5. Large Object Retention
    big_list = [0] * 10**6
    del big_list  # Freed in Python, but OS may not reclaim immediately

    # 6. External Resource Management
    f = open("file.txt", "w")
    f.write("Hello")
    f.close()  # Must close manually

## 25. Why is it important to use multithreading in certain applications?
- Multithreading allows a program to execute multiple threads (smaller units of a process) at the same time.
It is important in certain applications because:

**Improved Performance –** Multiple tasks can run in parallel, reducing total execution time.

**Better Resource Utilization –** Threads share the same memory space, avoiding overhead of separate processes.

**Responsive Applications – **Helps keep applications (like GUIs or servers) responsive while performing background tasks.

**Efficient I/O Handling –** Useful when tasks involve waiting for I/O (files, network, databases), so CPU time is not wasted.

Example:

    import threading
    import time

    def task(name):
        for i in range(3):
            print(f"{name} is running")
            time.sleep(1)

    # Creating threads
    t1 = threading.Thread(target=task, args=("Thread 1",))
    t2 = threading.Thread(target=task, args=("Thread 2",))

    # Starting threads
    t1.start()
    t2.start()

    # Waiting for threads to finish
    t1.join()
    t2.join()

    print("Both threads have finished!")


Here, two threads (Thread 1 and Thread 2) run at the same time.

Without multithreading, these tasks would run one after the other, taking longer.

With multithreading, both execute concurrently, improving efficiency

```python
# 6. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

import logging

logging.basicConfig(filename="math_errors.log", level=logging.ERROR,
                    format="%(levelname)s: %(message)s - Time: %(asctime)s")

def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        logging.error(f"Division by zero: attempted {a} / {b}")
        return None

x = 10
y = 0

result = divide(x, y)

if result is None:
    print("An error occurred. Check math_errors.log for details.")
else:
    print("Result:", result)
```

```python
# 12. Write a program that demonstrates using multiple except blocks to handle different types of exceptions.

try:
    num1 = 8
    num2 = 0

    result = num1 / num2
    print("Result:", result)

except ValueError:
    print("Error: Please enter valid integers only.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
# 14. Write a program that uses the logging module to log both informational and error messages.

import logging

logging.basicConfig(
    filename="app.log",               # Log file name
    level=logging.DEBUG,              # Log all levels DEBUG and above
    format="%(asctime)s - %(levelname)s - %(message)s")

logging.info("Program started successfully.")

try:
    num1 = 10
    num2 = 0  # This will cause division by zero
    result = num1 / num2
    logging.info(f"The result is {result}")

except ZeroDivisionError as e:
    logging.error(f"Error occurred: {e}")

logging.info("Program ended.")
```

```python
# 15. Write a Python program that prints the content of a file and handles the case when the file is empty.

filename = "notes.txt"

try:
    with open(filename, "r") as file:
        lines = file.readlines()  # Read all lines into a list
        if not lines:  # If the list is empty
            print(f"The file '{filename}' is empty.")
        else:
            print("Here are the contents of your notes:")
            for line in lines:
                print(line.strip())
except FileNotFoundError:
    print(f"Error: The file '{filename}' does not exist.")
```

```python
# 18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging
from logging.handlers import RotatingFileHandler

logger = logging.getLogger('MyLogger')
handler = RotatingFileHandler('my_log.log', maxBytes=1048576, backupCount=3)
logger.addHandler(handler)

logger.info("This is a log message.")
```

```python
# 21. Write a Python program that reads a file and prints the number of occurrences of a specific word?

word = "example"
count = 0

with open('example.txt', 'r') as file:
    for line in file:
        count += line.lower().count(word)

print(f"Word '{word}' occurred {count} times.")
```

```python
# 22. How can you check if a file is empty before attempting to read its contents?

import os

if os.stat('example.txt').st_size == 0:
    print("The file is empty.")
else:
    with open('example.txt', 'r') as file:
        print(file.read())
```
