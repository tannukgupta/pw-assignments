###Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

ANSWER:

AI (Artificial Intelligence), ML (Machine Learning), Deep Learning (DL), and Data Science (DS) are related but distinct fields within technology and data analysis. Here’s a clear breakdown of their key differences and relationships:

Artificial Intelligence (AI)

AI is the broadest field, focused on building machines or software that mimic human intelligence—thinking, reasoning, and decision-making without direct human programming.

AI includes systems that can use rules, logic, expert systems, and learning algorithms.

The scope of AI covers robotics, natural language processing, computer vision, and more.

- Machine Learning (ML)

ML is a subset of AI that deals explicitly with data-driven algorithms, enabling machines to learn from data and make predictions or decisions without being explicitly programmed for every scenario.

ML algorithms use statistical techniques for tasks like classification, regression, and clustering.

ML learning can be supervised, unsupervised, or reinforcement-based.

- Deep Learning (DL)
DL is a specialized subset of ML that uses artificial neural networks with many (deep) layers to analyze large and complex data sets.

DL excels at learning patterns in images, audio, and natural language by abstracting complex representations at multiple levels.

DL requires massive datasets and significant computational power for training, but can achieve state-of-the-art accuracy in tasks such as image and speech recognition.

- Data Science (DS)

DS is a multidisciplinary field focused on extracting insights, knowledge, and actionable information from data (structured or unstructured).

DS combines statistics, computer science, domain expertise, ML, and sometimes DL, to clean, analyze, visualize, and interpret data for business or research purposes.

DS is broader than ML, and while it often uses AI/ML/DL techniques, it also involves developing data pipelines, data wrangling, and business intelligence applications



###Question 2: What are the types of machine learning? Describe each with one real-world example.

ANSWER:

There are three main types of machine learning: supervised learning, unsupervised learning, and reinforcement learning. Each serves different purposes and is best explained with a real-world example.

- Supervised Learning-
The model is trained on a labeled dataset, where both input and output data are provided and the goal is to predict outputs for new inputs.

Example: Spam email detection—where the model learns from emails labeled as "spam" or "not spam" and then classifies new emails accordingly.

- Unsupervised Learning-
The model works on an unlabeled dataset and tries to find hidden patterns or intrinsic structures in the input data.

Example: Customer segmentation in marketing—where unsupervised algorithms group customers based on purchase behavior to identify distinct market segments, even though there are no pre-defined categories.

- Reinforcement Learning-
The model interacts with an environment and learns to make sequences of decisions, receiving rewards or penalties for actions taken.

Example: Training a robot to walk—where the robot learns which actions (steps) increase its stability and mobility through a reward system, improving its walking ability over time.




###Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

ANSWER:

Overfitting, underfitting, and the bias-variance tradeoff are crucial concepts for understanding and improving machine learning model performance.

- Overfitting-

Overfitting occurs when a model learns both the patterns and the noise in the training data, fitting it too closely.

Such a model performs very well on training data but poorly on new, unseen data, as it fails to generalize.

Overfitting is commonly associated with high variance and low bias.

- Underfitting-

Underfitting happens when a model is too simple to capture the underlying structure in the data.

The model performs poorly on both the training and the test data, missing key patterns and trends.

Underfitting is typically linked with high bias and low variance.

- Bias-Variance Tradeoff-

The bias-variance tradeoff describes the balance between two sources of model error: bias (error from erroneous assumptions) and variance (error from sensitivity to data fluctuations).

High bias models make strong assumptions and often underfit, while high variance models are overly sensitive to the training data and often overfit.

The goal is to find the optimal middle ground, where the model achieves a low total error by balancing bias and variance, thus generalizing well to unseen data.



###Question 4: What are outliers in a dataset, and list three common techniques for handling them.

ANSWER:

Outliers are data points in a dataset that deviate significantly from most other observations, often lying at the extreme ends of the value range. These unusual values can occur due to natural variation, measurement errors, or rare events, and can greatly influence statistical analyses or model performance.

Three Common Techniques for Handling Outliers-

- Z-score Method: This statistical approach identifies outliers by calculating how many standard deviations a data point is from the mean. Points with a Z-score above a chosen threshold (typically ±3) are classified as outliers and can be removed or treated.

- Interquartile Range (IQR) Method: Outliers are detected as values that fall below the lower bound (Q1 − 1.5 × IQR) or above the upper bound (Q3 + 1.5 × IQR), where Q1 and Q3 are the first and third quartiles, respectively.

- Clustering Algorithms: Density-based methods like DBSCAN or algorithms such as Isolation Forest are used to detect data points that do not fit into dense clusters or are easily isolated, marking them as outliers, especially in multidimensional datasets.

Choosing an appropriate technique depends on the dataset and the context of analysis, as well as whether outliers represent errors or true rare events


###Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.


ANSWER:

Handling missing values involves identifying gaps in the dataset and applying targeted strategies to achieve complete, reliable data for analysis or modeling. The process includes:

Detecting missing values using summary statistics or data visualization.

Understanding the pattern (random or systematic) and reason behind missingness.

Deciding whether to remove or impute missing values.

Applying appropriate imputation techniques relevant to the variable type.

Imputation Techniques

- For Numerical Data: Mean Imputation
Mean imputation replaces missing values with the average value of the observed entries in that feature.

This is straightforward and works well when the data is missing at random and is not highly skewed.

- For Categorical Data: Mode Imputation

Mode imputation substitutes missing values with the most frequently occurring (mode) category in that feature.

This preserves the distribution and is effective when one category dominates.

Other advanced techniques such as median imputation (for skewed numerical data), KNN imputation, and multiple imputations are also widely used in data science, but mean and mode remain the most common basic approaches.



```python
"""Question 6: Write a Python program that:
● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.
● Prints the class distribution. """

from sklearn.datasets import make_classification
from collections import Counter

# Create synthetic imbalanced dataset
data, labels = make_classification(n_classes=2, class_sep=2,
                                   weights=[0.9, 0.1],  # 90% for class 0, 10% for class 1
                                   n_samples=1000, n_features=10,
                                   n_informative=3, n_redundant=1,
                                   random_state=42)

# Calculate class distribution
class_distribution = Counter(labels)

# Print class distribution
print("Class distribution:")
for cls, count in class_distribution.items():
    print(f"Class {cls}: {count} samples")
```

```python
"""Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation."""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Plot histogram before imputation
df['Value'].hist(bins=30, alpha=0.7, color='blue')
plt.title('Histogram Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Fill missing values with the column mean
df['Value'].fillna(df['Value'].mean(), inplace=True)

# Plot histogram after imputation
df['Value'].hist(bins=30, alpha=0.7, color='green')
plt.title('Histogram After Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

###Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:
-  Missing ages,
-  Outliers in transaction amount,
-  A highly imbalanced target (fraud vs. non-fraud),
-  Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.


ANSWER:

A complete data preparation plan for your retail customer transaction dataset would include the following steps to address missing ages, outliers in transaction amount, target class imbalance, and categorical variables:

1. Handling Missing Data
For missing ages, use imputation techniques based on the pattern of missingness. If ages are missing at random, consider replacing them with the median or mean age. For more sophisticated handling, use regression imputation or machine learning models to predict missing values based on other features.

Always document the chosen imputation method, as this can impact downstream results.

2. Handling Outliers
For transaction amount outliers, visualize the distribution (with box plots or histograms) to identify extreme values.

Remove or cap outliers if they are likely to be errors. If legitimate, consider using robust scaling (e.g., using the interquartile range) or log-transforming transaction amounts to reduce their impact on model training.

Outlier handling improves data quality and the reliability of model predictions.

3. Handling Imbalanced Targets
For a highly imbalanced target (fraud vs. non-fraud), apply resampling methods. Use techniques like SMOTE (Synthetic Minority Oversampling Technique) or undersampling the majority class.

Ensemble methods (like balanced random forests or XGBoost with scale_pos_weight) and specialized loss functions are useful.

Also, consider evaluating with metrics like ROC-AUC and precision-recall curves, as accuracy can be misleading in imbalanced datasets.

4. Encoding Categorical Variables
For categorical variables such as payment method, use appropriate encoding:

For nominal variables (with no order), use one-hot encoding.

For ordinal variables, use ordinal encoding.

For high-cardinality categorical variables, consider target encoding or feature hashing to reduce dimensionality.

Ensure there is no data leakage by fitting encoding only on training data and applying to test data separately.

5. Data Standardization and Validation
Standardize and normalize numerical features after handling outliers to ensure consistent scale.

Split the data into training and test (or validation) sets to verify preprocessing steps.

Validate that all preprocessing did not introduce errors—run statistical checks, summarize distributions post-cleaning, and visualize if necessary.

Each step should be performed in a reproducible manner (using libraries and pipelines) and documented for transparency and production deployment
